```python
# -*- coding: utf-8 -*-
"""
================================================================================
EfficientNet-B2 Model + Training for Binary DR Detection (TWO-STAGE FINE-TUNING)
================================================================================

Two-Stage Training:
- Phase 1: Freeze backbone, train only classifier (faster convergence)
- Phase 2: Unfreeze backbone, fine-tune entire model (better final performance)

Model: EfficientNet-B2 @ 256x256 resolution
Uses cached preprocessed images from previous pipeline.

Author: [Your Name]
Date: 2025
================================================================================
"""

# ============================================================
# 1. Imports & Config
# ============================================================

import os
import random
import numpy as np
import pandas as pd
import cv2
import torch
from torch import nn, optim
from torch.utils.data import Dataset, DataLoader
from torch.cuda.amp import autocast, GradScaler
from torchvision import models, transforms
from sklearn.metrics import (
    accuracy_score, roc_auc_score,
    precision_score, recall_score,
    f1_score, confusion_matrix,
    classification_report, roc_curve,
    precision_recall_curve, average_precision_score
)
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import time
import copy
import warnings

warnings.filterwarnings('ignore')

print("=" * 70)
print("RETINAL CNN TRAINING PIPELINE (EfficientNet-B2 @ 256)")
print("=" * 70)


# ============================================================
# 2. Configuration
# ============================================================

class Config:
    """Data configuration."""
    DATA_DIR = "/content/drive/MyDrive/MULTIMODAL/datasets/EYEPACS/images/resized_train_cropped"
    SPLITS_DIR = "/content/drive/MyDrive/MULTIMODAL/datasets/EYEPACS/splits"
    CACHE_DIR = "/content/drive/MyDrive/MULTIMODAL/datasets/EYEPACS/preprocessed_cache_256"  # New cache for 256

    # ─────────────────────────────────────────────────────────
    # CHANGED: Image size from 224 to 256
    # ─────────────────────────────────────────────────────────
    IMAGE_SIZE = 256

    # ─────────────────────────────────────────────────────────
    # CHANGED: Reduced batch size for larger model
    # ─────────────────────────────────────────────────────────
    BATCH_SIZE = 16  # Reduced from 32 due to larger model

    NUM_WORKERS = 2
    SEED = 42


class ModelConfig:
    """Model and training configuration."""

    # ─────────────────────────────────────────────────────────
    # CHANGED: EfficientNet-B2 instead of B0
    # ─────────────────────────────────────────────────────────
    MODEL_NAME = 'efficientnet_b2'

    PRETRAINED = True
    NUM_EPOCHS = 40
    LR = 1e-4
    WEIGHT_DECAY = 1e-5
    DROPOUT = 0.3
    POS_WEIGHT = 2.0
    DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    SAVE_DIR = "/content/drive/MyDrive/MULTIMODAL/models"

    # ─────────────────────────────────────────────────────────
    # CHANGED: Model save name
    # ─────────────────────────────────────────────────────────
    MODEL_SAVE_NAME = "efficientnet_b2_binary_dr_twostage.pth"

    # Preprocessing options (must match cache)
    USE_BEN_GRAHAM = True
    USE_CLAHE = False
    USE_CIRCLE_CROP = True

    # Optimization options
    USE_MIXED_PRECISION = True
    USE_CACHED_IMAGES = True

    # Early stopping
    EARLY_STOPPING_PATIENCE = 7
    EARLY_STOPPING_MIN_DELTA = 0.001

    # Two-Stage Fine-Tuning Parameters
    FREEZE_EPOCHS = 5
    UNFREEZE_LR_FACTOR = 0.5
    FREEZE_BN = False


def seed_everything(seed: int = 42):
    """Set random seeds for reproducibility."""
    os.environ['PYTHONHASHSEED'] = str(seed)
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
    print(f"[INFO] Random seed set to {seed}")


# ============================================================
# 3. Image Preprocessing Functions (for caching)
# ============================================================

def crop_black(img, tolerance=7):
    """Remove black borders from fundus image."""
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    mask = gray > tolerance
    coords = np.argwhere(mask)
    if coords.size == 0:
        return img
    y0, x0 = coords.min(0)
    y1, x1 = coords.max(0)
    return img[y0:y1+1, x0:x1+1]


def apply_clahe(img, clip_limit=2.0, tile_grid_size=(8, 8)):
    """Apply CLAHE enhancement to image."""
    lab = cv2.cvtColor(img, cv2.COLOR_RGB2LAB)
    clahe = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=tile_grid_size)
    lab[:, :, 0] = clahe.apply(lab[:, :, 0])
    return cv2.cvtColor(lab, cv2.COLOR_LAB2RGB)


def circle_crop(img):
    """Apply circular mask to remove corners."""
    h, w = img.shape[:2]
    r = min(h, w) // 2
    mask = np.zeros((h, w), np.uint8)
    cv2.circle(mask, (w // 2, h // 2), r, 255, -1)
    return cv2.bitwise_and(img, img, mask=mask)


def preprocess_image_for_cache(path, image_size=256,
                                use_clahe=False,
                                use_circle_crop=True,
                                use_ben_graham=True):
    """Preprocess image and return numpy array for caching."""
    img = cv2.imread(path)
    if img is None:
        return None
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = crop_black(img)

    # Resize a bit larger to keep detail
    base_size = max(image_size, 288)  # Adjusted for 256
    img = cv2.resize(img, (base_size, base_size))

    if use_ben_graham:
        blur = cv2.GaussianBlur(img, (0, 0), 10)
        img = cv2.addWeighted(img, 4, blur, -4, 128)
    elif use_clahe:
        img = apply_clahe(img)

    if use_circle_crop:
        img = circle_crop(img)

    img = cv2.resize(img, (image_size, image_size))
    img = np.clip(img, 0, 255).astype(np.uint8)
    return img


# ============================================================
# 4. Cache Functions
# ============================================================

def create_preprocessed_cache(config, model_cfg):
    """Preprocess all images and save to cache directory."""
    print("\n" + "=" * 60)
    print(f"CREATING PREPROCESSED IMAGE CACHE ({config.IMAGE_SIZE}x{config.IMAGE_SIZE})")
    print("=" * 60)

    train_df = pd.read_csv(f"{config.SPLITS_DIR}/retinal_train_images.csv")
    val_df = pd.read_csv(f"{config.SPLITS_DIR}/retinal_val_images.csv")
    test_df = pd.read_csv(f"{config.SPLITS_DIR}/retinal_test_images.csv")

    splits = {'train': train_df, 'val': val_df, 'test': test_df}

    for split_name, df in splits.items():
        split_cache_dir = os.path.join(config.CACHE_DIR, split_name)
        os.makedirs(split_cache_dir, exist_ok=True)

        print(f"\nProcessing {split_name} set: {len(df)} images")
        existing_files = set(os.listdir(split_cache_dir))

        cached_count, failed_count, skipped_count = 0, 0, 0

        for idx, row in tqdm(df.iterrows(), total=len(df), desc=f"Caching {split_name}"):
            image_name = row['image']
            cache_filename = f"{image_name}.npy"
            cache_path = os.path.join(split_cache_dir, cache_filename)

            if cache_filename in existing_files:
                skipped_count += 1
                continue

            img_path = os.path.join(config.DATA_DIR, image_name + ".jpeg")
            preprocessed = preprocess_image_for_cache(
                img_path, image_size=config.IMAGE_SIZE,
                use_clahe=model_cfg.USE_CLAHE,
                use_circle_crop=model_cfg.USE_CIRCLE_CROP,
                use_ben_graham=model_cfg.USE_BEN_GRAHAM
            )

            if preprocessed is not None:
                np.save(cache_path, preprocessed)
                cached_count += 1
            else:
                failed_count += 1

        print(f"  Cached: {cached_count}, Skipped: {skipped_count}, Failed: {failed_count}")

    print("\n✓ Cache creation complete!")
    return True


def verify_cache(config):
    """Verify that cache exists and has expected files."""
    splits = ['train', 'val', 'test']
    cache_valid = True

    for split in splits:
        split_dir = os.path.join(config.CACHE_DIR, split)
        if not os.path.exists(split_dir):
            print(f"[WARNING] Cache directory missing: {split_dir}")
            cache_valid = False
            continue

        csv_path = os.path.join(config.SPLITS_DIR, f"retinal_{split}_images.csv")
        df = pd.read_csv(csv_path)

        cached_files = set(os.listdir(split_dir))
        expected_files = set(f"{row['image']}.npy" for _, row in df.iterrows())

        missing = expected_files - cached_files
        if len(missing) > 0:
            print(f"[WARNING] {split}: {len(missing)} files missing from cache")
            cache_valid = False
        else:
            print(f"[OK] {split}: {len(cached_files)} cached files verified")

    return cache_valid


# ============================================================
# 5. Data Augmentation Transforms
# ============================================================

def get_train_transform(image_size=256):
    """Get training augmentation transforms."""
    return transforms.Compose([
        transforms.RandomResizedCrop(image_size, scale=(0.8, 1.0)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(15),
        transforms.ColorJitter(brightness=0.2, contrast=0.2),
        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                             std=[0.229, 0.224, 0.225])
    ])


def get_val_test_transform(image_size=256):
    """Get validation/test transforms (normalization only)."""
    return transforms.Compose([
        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                             std=[0.229, 0.224, 0.225])
    ])


# ============================================================
# 6. Dataset Class
# ============================================================

class CachedRetinalDataset(Dataset):
    """Dataset that loads preprocessed images from cache."""

    def __init__(self, df, cache_dir, is_training=True, image_size=256):
        self.df = df.reset_index(drop=True)
        self.cache_dir = cache_dir
        self.is_training = is_training
        self.size = image_size
        self.transform = get_train_transform(image_size) if is_training else get_val_test_transform(image_size)

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        image_name = row['image']
        label = float(row['patient_label'])

        cache_path = os.path.join(self.cache_dir, f"{image_name}.npy")

        try:
            img = np.load(cache_path)
            img = img.astype(np.float32) / 255.0
            img = torch.from_numpy(img).permute(2, 0, 1)
        except:
            img = torch.zeros(3, self.size, self.size)

        img = self.transform(img)

        return {
            'image': img,
            'label': torch.tensor(label, dtype=torch.float32)
        }


# ============================================================
# 7. Model Definition (Updated for B0, B1, B2, B3)
# ============================================================

class BinaryDRClassifier(nn.Module):
    """Binary Diabetic Retinopathy Classifier using EfficientNet backbone."""

    def __init__(self, model_name='efficientnet_b2', pretrained=True, dropout=0.3):
        super().__init__()

        self.model_name = model_name

        # ─────────────────────────────────────────────────────────
        # Support multiple EfficientNet variants
        # ─────────────────────────────────────────────────────────
        if model_name == 'efficientnet_b0':
            weights = models.EfficientNet_B0_Weights.IMAGENET1K_V1 if pretrained else None
            self.backbone = models.efficientnet_b0(weights=weights)
        elif model_name == 'efficientnet_b1':
            weights = models.EfficientNet_B1_Weights.IMAGENET1K_V1 if pretrained else None
            self.backbone = models.efficientnet_b1(weights=weights)
        elif model_name == 'efficientnet_b2':
            weights = models.EfficientNet_B2_Weights.IMAGENET1K_V1 if pretrained else None
            self.backbone = models.efficientnet_b2(weights=weights)
        elif model_name == 'efficientnet_b3':
            weights = models.EfficientNet_B3_Weights.IMAGENET1K_V1 if pretrained else None
            self.backbone = models.efficientnet_b3(weights=weights)
        else:
            raise ValueError(f'Model {model_name} not supported. Use efficientnet_b0, b1, b2, or b3.')

        # Get feature dimension from the backbone
        n = self.backbone.classifier[1].in_features
        self.backbone.classifier = nn.Identity()
        self.feature_dim = n

        # Classifier head
        self.classifier = nn.Sequential(
            nn.Dropout(dropout),
            nn.Linear(n, 256),
            nn.ReLU(),
            nn.BatchNorm1d(256),
            nn.Dropout(dropout * 0.5),
            nn.Linear(256, 1)
        )

    def extract_features(self, x):
        """Extract features from backbone for multimodal fusion."""
        return self.backbone(x)

    def forward(self, x):
        """Forward pass returning logits."""
        f = self.backbone(x)
        return self.classifier(f)


# ============================================================
# 8. Freeze/Unfreeze Helper Functions
# ============================================================

def freeze_backbone(model, freeze_bn=False):
    """Freeze all backbone parameters."""
    for param in model.backbone.parameters():
        param.requires_grad = False

    if not freeze_bn:
        for module in model.backbone.modules():
            if isinstance(module, (nn.BatchNorm2d, nn.BatchNorm1d)):
                for param in module.parameters():
                    param.requires_grad = True

    trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
    total = sum(p.numel() for p in model.parameters())

    print(f"[FREEZE] Backbone frozen. Trainable: {trainable:,} / {total:,} params "
          f"({100*trainable/total:.1f}%)")


def unfreeze_backbone(model):
    """Unfreeze all backbone parameters for full fine-tuning."""
    for param in model.backbone.parameters():
        param.requires_grad = True

    trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
    total = sum(p.numel() for p in model.parameters())

    print(f"[UNFREEZE] Backbone unfrozen. Trainable: {trainable:,} / {total:,} params "
          f"({100*trainable/total:.1f}%)")


def get_trainable_params(model):
    """Get only trainable parameters for optimizer."""
    return [p for p in model.parameters() if p.requires_grad]


def create_model(cfg):
    """Create and initialize the model."""
    model = BinaryDRClassifier(
        cfg.MODEL_NAME,
        cfg.PRETRAINED,
        cfg.DROPOUT
    )
    model.to(cfg.DEVICE)
    os.makedirs(cfg.SAVE_DIR, exist_ok=True)

    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

    print(f"\n[INFO] Model: {cfg.MODEL_NAME}")
    print(f"[INFO] Device: {cfg.DEVICE}")
    print(f"[INFO] Total parameters: {total_params:,}")
    print(f"[INFO] Trainable parameters: {trainable_params:,}")
    print(f"[INFO] Feature dimension: {model.feature_dim}")

    return model


# ============================================================
# 9. Early Stopping
# ============================================================

class EarlyStopping:
    """Early stopping to prevent overfitting."""

    def __init__(self, patience=7, min_delta=0.001, mode='max'):
        self.patience = patience
        self.min_delta = min_delta
        self.mode = mode
        self.counter = 0
        self.best_score = None
        self.early_stop = False
        self.best_epoch = 0

    def __call__(self, score, epoch):
        if self.best_score is None:
            self.best_score = score
            self.best_epoch = epoch
            return False

        if self._is_improvement(score):
            self.best_score = score
            self.best_epoch = epoch
            self.counter = 0
        else:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True

        return self.early_stop

    def _is_improvement(self, score):
        if self.mode == 'max':
            return score > self.best_score + self.min_delta
        return score < self.best_score - self.min_delta

    def reset(self):
        """Reset early stopping counter (useful when unfreezing)."""
        self.counter = 0


# ============================================================
# 10. Training Functions
# ============================================================

def train_one_epoch(model, loader, criterion, optimizer, device, scaler=None, use_amp=True):
    """Train for one epoch with optional mixed precision."""
    model.train()
    total_loss = 0
    preds = []
    labels = []

    for batch in tqdm(loader, desc="Train", leave=False):
        x = batch['image'].to(device)
        y = batch['label'].to(device).unsqueeze(1)

        optimizer.zero_grad()

        if use_amp and scaler is not None:
            with autocast():
                out = model(x)
                loss = criterion(out, y)

            scaler.scale(loss).backward()
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            scaler.step(optimizer)
            scaler.update()
        else:
            out = model(x)
            loss = criterion(out, y)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()

        total_loss += loss.item() * x.size(0)

        with torch.no_grad():
            probs = torch.sigmoid(out.float()).cpu().numpy().ravel()
        preds.extend(probs)
        labels.extend(y.cpu().numpy().ravel())

    epoch_loss = total_loss / len(loader.dataset)
    preds = np.array(preds)
    labels = np.array(labels)

    auc = roc_auc_score(labels, preds)
    acc = accuracy_score(labels, (preds > 0.5).astype(int))

    return {'loss': epoch_loss, 'auc': auc, 'acc': acc}


def validate(model, loader, criterion, device, use_amp=True):
    """Validate the model with optional mixed precision."""
    model.eval()
    total_loss = 0
    preds = []
    labels = []

    with torch.no_grad():
        for batch in tqdm(loader, desc="Val", leave=False):
            x = batch['image'].to(device)
            y = batch['label'].to(device).unsqueeze(1)

            if use_amp:
                with autocast():
                    out = model(x)
                    loss = criterion(out, y)
            else:
                out = model(x)
                loss = criterion(out, y)

            total_loss += loss.item() * x.size(0)

            probs = torch.sigmoid(out.float()).cpu().numpy().ravel()
            preds.extend(probs)
            labels.extend(y.cpu().numpy().ravel())

    epoch_loss = total_loss / len(loader.dataset)
    preds = np.array(preds)
    labels = np.array(labels)

    auc = roc_auc_score(labels, preds)
    acc = accuracy_score(labels, (preds > 0.5).astype(int))

    return {'loss': epoch_loss, 'auc': auc, 'acc': acc}


# ============================================================
# 11. Two-Stage Training Loop
# ============================================================

def train_model_two_stage(model, train_loader, val_loader, cfg):
    """
    Two-stage training loop:
    - Phase 1: Backbone frozen, train classifier only
    - Phase 2: Full fine-tuning with lower LR
    """

    pos_weight = torch.tensor([cfg.POS_WEIGHT], device=cfg.DEVICE)
    criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weight)

    scaler = GradScaler() if cfg.USE_MIXED_PRECISION and cfg.DEVICE.type == 'cuda' else None
    use_amp = cfg.USE_MIXED_PRECISION and cfg.DEVICE.type == 'cuda'

    early_stopping = EarlyStopping(
        patience=cfg.EARLY_STOPPING_PATIENCE,
        min_delta=cfg.EARLY_STOPPING_MIN_DELTA,
        mode='max'
    )

    best_auc = 0.0
    best_weights = None
    history = {
        'train_loss': [], 'train_auc': [], 'train_acc': [],
        'val_loss': [], 'val_auc': [], 'val_acc': [],
        'lr': [], 'phase': []
    }

    print("\n" + "=" * 70)
    print("TWO-STAGE TRAINING STARTED")
    print("=" * 70)
    print(f"Model: {cfg.MODEL_NAME}")
    print(f"Image Size: {Config.IMAGE_SIZE}x{Config.IMAGE_SIZE}")
    print(f"Phase 1: Epochs 1-{cfg.FREEZE_EPOCHS} (Backbone FROZEN, LR={cfg.LR})")
    print(f"Phase 2: Epochs {cfg.FREEZE_EPOCHS+1}-{cfg.NUM_EPOCHS} (Backbone UNFROZEN, LR={cfg.LR * cfg.UNFREEZE_LR_FACTOR})")
    print(f"Pos weight: {cfg.POS_WEIGHT}")
    print(f"Batch size: {Config.BATCH_SIZE}")
    print(f"Mixed precision: {use_amp}")
    print(f"Early stopping patience: {cfg.EARLY_STOPPING_PATIENCE}")
    print(f"Device: {cfg.DEVICE}")
    print("=" * 70)

    # ─────────────────────────────────────────────────────────
    # PHASE 1: Freeze backbone
    # ─────────────────────────────────────────────────────────
    print("\n" + "─" * 50)
    print("PHASE 1: Training classifier with frozen backbone")
    print("─" * 50)

    freeze_backbone(model, freeze_bn=cfg.FREEZE_BN)

    optimizer = optim.AdamW(
        get_trainable_params(model),
        lr=cfg.LR,
        weight_decay=cfg.WEIGHT_DECAY
    )

    scheduler = optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='max', patience=2, factor=0.5, min_lr=1e-6
    )

    start_time = time.time()

    for epoch in range(1, cfg.NUM_EPOCHS + 1):
        epoch_start = time.time()

        # ─────────────────────────────────────────────────────
        # PHASE 2: Unfreeze at FREEZE_EPOCHS + 1
        # ─────────────────────────────────────────────────────
        if epoch == cfg.FREEZE_EPOCHS + 1:
            print("\n" + "─" * 50)
            print("PHASE 2: Fine-tuning entire model (backbone unfrozen)")
            print("─" * 50)

            unfreeze_backbone(model)

            new_lr = cfg.LR * cfg.UNFREEZE_LR_FACTOR
            optimizer = optim.AdamW(
                model.parameters(),
                lr=new_lr,
                weight_decay=cfg.WEIGHT_DECAY
            )

            scheduler = optim.lr_scheduler.ReduceLROnPlateau(
                optimizer, mode='max', patience=3, factor=0.5, min_lr=1e-6
            )

            early_stopping.reset()
            print(f"[INFO] LR reduced to {new_lr:.2e} for fine-tuning")
            print(f"[INFO] Early stopping counter reset")

        current_lr = optimizer.param_groups[0]['lr']
        phase = 1 if epoch <= cfg.FREEZE_EPOCHS else 2

        print(f"\nEpoch {epoch}/{cfg.NUM_EPOCHS} [Phase {phase}] (LR: {current_lr:.2e})")
        print("-" * 40)

        t = train_one_epoch(model, train_loader, criterion, optimizer,
                           cfg.DEVICE, scaler, use_amp)

        v = validate(model, val_loader, criterion, cfg.DEVICE, use_amp)

        scheduler.step(v['auc'])

        history['train_loss'].append(t['loss'])
        history['train_auc'].append(t['auc'])
        history['train_acc'].append(t['acc'])
        history['val_loss'].append(v['loss'])
        history['val_auc'].append(v['auc'])
        history['val_acc'].append(v['acc'])
        history['lr'].append(current_lr)
        history['phase'].append(phase)

        epoch_time = time.time() - epoch_start

        print(f"Train  Loss: {t['loss']:.4f} | AUC: {t['auc']:.4f} | Acc: {t['acc']:.4f}")
        print(f"Val    Loss: {v['loss']:.4f} | AUC: {v['auc']:.4f} | Acc: {v['acc']:.4f}")
        print(f"Time: {epoch_time:.1f}s")

        if v['auc'] > best_auc:
            best_auc = v['auc']
            best_weights = copy.deepcopy(model.state_dict())

            checkpoint = {
                'epoch': epoch,
                'phase': phase,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'val_auc': v['auc'],
                'val_acc': v['acc'],
                'feature_dim': model.feature_dim,
                'config': {
                    'model_name': cfg.MODEL_NAME,
                    'image_size': Config.IMAGE_SIZE,
                    'dropout': cfg.DROPOUT,
                    'use_ben_graham': cfg.USE_BEN_GRAHAM,
                    'use_clahe': cfg.USE_CLAHE,
                    'use_circle_crop': cfg.USE_CIRCLE_CROP,
                    'freeze_epochs': cfg.FREEZE_EPOCHS,
                    'unfreeze_lr_factor': cfg.UNFREEZE_LR_FACTOR
                }
            }
            torch.save(checkpoint, os.path.join(cfg.SAVE_DIR, cfg.MODEL_SAVE_NAME))
            print(f"  ✅ Saved best model (Phase {phase}, AUC: {v['auc']:.4f})")

        if phase == 2:
            if early_stopping(v['auc'], epoch):
                print(f"\n⚠️ Early stopping triggered at epoch {epoch}")
                print(f"   Best epoch: {early_stopping.best_epoch} with AUC: {early_stopping.best_score:.4f}")
                break

    total_time = time.time() - start_time

    model.load_state_dict(best_weights)

    print("\n" + "=" * 70)
    print("TWO-STAGE TRAINING COMPLETED")
    print("=" * 70)
    print(f"Total time: {total_time:.1f}s ({total_time/60:.2f} min)")
    print(f"Epochs trained: {len(history['train_loss'])}")
    print(f"  Phase 1 (frozen): {min(cfg.FREEZE_EPOCHS, len(history['train_loss']))} epochs")
    print(f"  Phase 2 (fine-tuned): {max(0, len(history['train_loss']) - cfg.FREEZE_EPOCHS)} epochs")
    print(f"Best Validation AUC: {best_auc:.4f}")

    return model, history


# ============================================================
# 12. Threshold Tuning
# ============================================================

def tune_threshold(labels, probs):
    """Find optimal threshold for F1 score."""
    labels = np.array(labels)
    probs = np.array(probs)

    best_thr = 0.5
    best_f1 = 0.0

    print("\n" + "-" * 70)
    print(f"{'Threshold':<12} {'Precision':<12} {'Recall':<12} {'F1-Score':<12}")
    print("-" * 70)

    for thr in np.linspace(0.1, 0.9, 17):
        preds_bin = (probs >= thr).astype(int)
        p = precision_score(labels, preds_bin, zero_division=0)
        r = recall_score(labels, preds_bin, zero_division=0)
        f1 = f1_score(labels, preds_bin, zero_division=0)

        print(f"{thr:<12.2f} {p:<12.3f} {r:<12.3f} {f1:<12.3f}")

        if f1 > best_f1:
            best_f1 = f1
            best_thr = thr

    print("-" * 70)
    print(f"\n✓ Best F1 threshold: {best_thr:.2f} (F1 = {best_f1:.3f})")

    return best_thr, best_f1


# ============================================================
# 13. Evaluation
# ============================================================

def evaluate_model(model, loader, device, use_amp=True):
    """Comprehensive model evaluation with threshold tuning."""
    model.eval()
    probs = []
    labels = []

    with torch.no_grad():
        for batch in tqdm(loader, desc="Test"):
            x = batch['image'].to(device)
            y = batch['label'].to(device)

            if use_amp:
                with autocast():
                    out = model(x)
            else:
                out = model(x)

            p = torch.sigmoid(out.float()).cpu().numpy().ravel()
            probs.extend(p)
            labels.extend(y.cpu().numpy().ravel())

    probs = np.array(probs)
    labels = np.array(labels)

    binary_default = (probs >= 0.5).astype(int)

    print("\n" + "=" * 60)
    print("EVALUATION RESULTS")
    print("=" * 60)

    print("\n--- Results with Default Threshold (0.50) ---")
    print("\nClassification Report:")
    print(classification_report(labels, binary_default, target_names=['No DR', 'DR']))

    print("\n--- Threshold Optimization ---")
    best_thr, best_f1 = tune_threshold(labels, probs)

    binary_best = (probs >= best_thr).astype(int)

    print(f"\n--- Results with Optimal Threshold ({best_thr:.2f}) ---")
    print("\nClassification Report:")
    print(classification_report(labels, binary_best, target_names=['No DR', 'DR']))

    auc = roc_auc_score(labels, probs)

    acc_default = accuracy_score(labels, binary_default)
    prec_default = precision_score(labels, binary_default, zero_division=0)
    rec_default = recall_score(labels, binary_default, zero_division=0)
    f1_default = f1_score(labels, binary_default, zero_division=0)

    acc_optimal = accuracy_score(labels, binary_best)
    prec_optimal = precision_score(labels, binary_best, zero_division=0)
    rec_optimal = recall_score(labels, binary_best, zero_division=0)
    f1_optimal = f1_score(labels, binary_best, zero_division=0)

    print("\n" + "=" * 60)
    print("THRESHOLD COMPARISON")
    print("=" * 60)

    print(f"""
┌─────────────────┬──────────────────┬──────────────────┬─────────────┐
│ Metric          │ Threshold = 0.50 │ Threshold = {best_thr:.2f} │ Change      │
├─────────────────┼──────────────────┼──────────────────┼─────────────┤
│ Accuracy        │ {acc_default:.4f}           │ {acc_optimal:.4f}           │ {acc_optimal - acc_default:+.4f}      │
│ Precision       │ {prec_default:.4f}           │ {prec_optimal:.4f}           │ {prec_optimal - prec_default:+.4f}      │
│ Recall          │ {rec_default:.4f}           │ {rec_optimal:.4f}           │ {rec_optimal - rec_default:+.4f}      │
│ F1-Score        │ {f1_default:.4f}           │ {f1_optimal:.4f}           │ {f1_optimal - f1_default:+.4f}      │
└─────────────────┴──────────────────┴──────────────────┴─────────────┘

AUC-ROC: {auc:.4f} (threshold-independent)
""")

    # Visualizations
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    fig.suptitle(f'Test Set Evaluation ({ModelConfig.MODEL_NAME} @ {Config.IMAGE_SIZE})',
                 fontsize=14, fontweight='bold')

    cm_default = confusion_matrix(labels, binary_default)
    sns.heatmap(cm_default, annot=True, fmt='d', cmap='Blues', ax=axes[0],
                xticklabels=['No DR', 'DR'], yticklabels=['No DR', 'DR'])
    axes[0].set_title('Confusion Matrix (thr=0.50)')
    axes[0].set_xlabel('Predicted')
    axes[0].set_ylabel('Actual')

    cm_optimal = confusion_matrix(labels, binary_best)
    sns.heatmap(cm_optimal, annot=True, fmt='d', cmap='Greens', ax=axes[1],
                xticklabels=['No DR', 'DR'], yticklabels=['No DR', 'DR'])
    axes[1].set_title(f'Confusion Matrix (thr={best_thr:.2f})')
    axes[1].set_xlabel('Predicted')
    axes[1].set_ylabel('Actual')

    fpr, tpr, _ = roc_curve(labels, probs)
    axes[2].plot(fpr, tpr, 'b-', linewidth=2, label=f'ROC (AUC={auc:.3f})')
    axes[2].plot([0, 1], [0, 1], 'k--')
    axes[2].set_xlabel('False Positive Rate')
    axes[2].set_ylabel('True Positive Rate')
    axes[2].set_title('ROC Curve')
    axes[2].legend()
    axes[2].grid(alpha=0.3)

    plt.tight_layout()
    plt.show()

    print(f"\nTest AUC: {auc:.4f}")

    return {
        'auc': auc,
        'optimal_threshold': best_thr,
        'accuracy_default': acc_default,
        'accuracy_optimal': acc_optimal,
        'precision_default': prec_default,
        'precision_optimal': prec_optimal,
        'recall_default': rec_default,
        'recall_optimal': rec_optimal,
        'f1_default': f1_default,
        'f1_optimal': f1_optimal
    }


# ============================================================
# 14. Plotting Functions
# ============================================================

def plot_training_history_two_stage(history, freeze_epochs, save_path=None):
    """Plot training curves with phase markers."""
    epochs = range(1, len(history['train_loss']) + 1)

    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    fig.suptitle(f'Two-Stage Training History ({ModelConfig.MODEL_NAME} @ {Config.IMAGE_SIZE})',
                 fontsize=14, fontweight='bold')

    # Loss
    axes[0, 0].plot(epochs, history['train_loss'], 'b-', label='Train')
    axes[0, 0].plot(epochs, history['val_loss'], 'r-', label='Validation')
    if freeze_epochs < len(epochs):
        axes[0, 0].axvline(x=freeze_epochs + 0.5, color='green', linestyle='--',
                          label='Unfreeze', alpha=0.7)
    axes[0, 0].set_xlabel('Epoch')
    axes[0, 0].set_ylabel('Loss')
    axes[0, 0].set_title('Training vs Validation Loss')
    axes[0, 0].legend()
    axes[0, 0].grid(alpha=0.3)

    # AUC
    axes[0, 1].plot(epochs, history['train_auc'], 'b-', label='Train')
    axes[0, 1].plot(epochs, history['val_auc'], 'r-', label='Validation')
    if freeze_epochs < len(epochs):
        axes[0, 1].axvline(x=freeze_epochs + 0.5, color='green', linestyle='--',
                          label='Unfreeze', alpha=0.7)
    axes[0, 1].set_xlabel('Epoch')
    axes[0, 1].set_ylabel('AUC-ROC')
    axes[0, 1].set_title('Training vs Validation AUC')
    axes[0, 1].legend()
    axes[0, 1].grid(alpha=0.3)

    # Accuracy
    axes[1, 0].plot(epochs, history['train_acc'], 'b-', label='Train')
    axes[1, 0].plot(epochs, history['val_acc'], 'r-', label='Validation')
    if freeze_epochs < len(epochs):
        axes[1, 0].axvline(x=freeze_epochs + 0.5, color='green', linestyle='--',
                          label='Unfreeze', alpha=0.7)
    axes[1, 0].set_xlabel('Epoch')
    axes[1, 0].set_ylabel('Accuracy')
    axes[1, 0].set_title('Training vs Validation Accuracy')
    axes[1, 0].legend()
    axes[1, 0].grid(alpha=0.3)

    # Learning Rate
    axes[1, 1].plot(epochs, history['lr'], 'g-')
    if freeze_epochs < len(epochs):
        axes[1, 1].axvline(x=freeze_epochs + 0.5, color='green', linestyle='--',
                          label='Unfreeze', alpha=0.7)
    axes[1, 1].set_xlabel('Epoch')
    axes[1, 1].set_ylabel('Learning Rate')
    axes[1, 1].set_title('Learning Rate Schedule')
    axes[1, 1].set_yscale('log')
    axes[1, 1].legend()
    axes[1, 1].grid(alpha=0.3)

    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
        print(f"Saved: {save_path}")

    plt.show()


# ============================================================
# 15. Main Execution
# ============================================================

if __name__ == "__main__":

    # Initialize
    config = Config()
    model_cfg = ModelConfig()
    seed_everything(config.SEED)

    # Print configuration
    print("\n" + "=" * 60)
    print("EXPERIMENT CONFIGURATION")
    print("=" * 60)
    print(f"Model:       {model_cfg.MODEL_NAME}")
    print(f"Image Size:  {config.IMAGE_SIZE}x{config.IMAGE_SIZE}")
    print(f"Batch Size:  {config.BATCH_SIZE}")
    print(f"Cache Dir:   {config.CACHE_DIR}")
    print("=" * 60)

    # ─────────────────────────────────────────────────────────
    # Step 1: Check/Create Cache (new cache for 256 resolution)
    # ─────────────────────────────────────────────────────────
    if model_cfg.USE_CACHED_IMAGES:
        print("\n" + "-" * 40)
        print(f"Checking preprocessed image cache ({config.IMAGE_SIZE}x{config.IMAGE_SIZE})...")
        print("-" * 40)

        cache_exists = os.path.exists(config.CACHE_DIR)

        if not cache_exists or not verify_cache(config):
            print(f"\n[INFO] Cache for {config.IMAGE_SIZE}x{config.IMAGE_SIZE} missing. Creating cache...")
            create_preprocessed_cache(config, model_cfg)
            print("\n[INFO] Verifying cache...")
            verify_cache(config)
        else:
            print("\n[INFO] Cache verified successfully!")

    # ─────────────────────────────────────────────────────────
    # Step 2: Load CSV splits
    # ─────────────────────────────────────────────────────────
    print("\n" + "-" * 40)
    print("Loading data splits...")
    print("-" * 40)

    train_df = pd.read_csv(f"{config.SPLITS_DIR}/retinal_train_images.csv")
    val_df = pd.read_csv(f"{config.SPLITS_DIR}/retinal_val_images.csv")
    test_df = pd.read_csv(f"{config.SPLITS_DIR}/retinal_test_images.csv")

    print(f"Train: {len(train_df)} images")
    print(f"Val:   {len(val_df)} images")
    print(f"Test:  {len(test_df)} images")

    print(f"\nTrain class distribution:")
    print(train_df['patient_label'].value_counts().to_dict())

    # ─────────────────────────────────────────────────────────
    # Step 3: Create DataLoaders
    # ─────────────────────────────────────────────────────────
    print("\n" + "-" * 40)
    print("Creating DataLoaders...")
    print("-" * 40)

    train_dataset = CachedRetinalDataset(
        train_df,
        cache_dir=os.path.join(config.CACHE_DIR, 'train'),
        is_training=True,
        image_size=config.IMAGE_SIZE
    )

    val_dataset = CachedRetinalDataset(
        val_df,
        cache_dir=os.path.join(config.CACHE_DIR, 'val'),
        is_training=False,
        image_size=config.IMAGE_SIZE
    )

    test_dataset = CachedRetinalDataset(
        test_df,
        cache_dir=os.path.join(config.CACHE_DIR, 'test'),
        is_training=False,
        image_size=config.IMAGE_SIZE
    )

    train_loader = DataLoader(
        train_dataset, batch_size=config.BATCH_SIZE, shuffle=True,
        num_workers=config.NUM_WORKERS, drop_last=True, pin_memory=True
    )

    val_loader = DataLoader(
        val_dataset, batch_size=config.BATCH_SIZE, shuffle=False,
        num_workers=config.NUM_WORKERS, pin_memory=True
    )

    test_loader = DataLoader(
        test_dataset, batch_size=config.BATCH_SIZE, shuffle=False,
        num_workers=config.NUM_WORKERS, pin_memory=True
    )

    print(f"Train batches: {len(train_loader)}")
    print(f"Val batches:   {len(val_loader)}")
    print(f"Test batches:  {len(test_loader)}")

    # ─────────────────────────────────────────────────────────
    # Step 4: Create and train model (TWO-STAGE)
    # ─────────────────────────────────────────────────────────
    model = create_model(model_cfg)

    trained_model, history = train_model_two_stage(model, train_loader, val_loader, model_cfg)

    # ─────────────────────────────────────────────────────────
    # Step 5: Evaluate on test set
    # ─────────────────────────────────────────────────────────
    use_amp = model_cfg.USE_MIXED_PRECISION and model_cfg.DEVICE.type == 'cuda'
    test_results = evaluate_model(trained_model, test_loader, model_cfg.DEVICE, use_amp)

    # ─────────────────────────────────────────────────────────
    # Step 6: Plot training history
    # ─────────────────────────────────────────────────────────
    plot_training_history_two_stage(
        history,
        freeze_epochs=model_cfg.FREEZE_EPOCHS,
        save_path=os.path.join(model_cfg.SAVE_DIR, f"retinal_training_history_{model_cfg.MODEL_NAME}.png")
    )

    # ─────────────────────────────────────────────────────────
    # Step 7: Save final results
    # ─────────────────────────────────────────────────────────
    checkpoint_path = os.path.join(model_cfg.SAVE_DIR, model_cfg.MODEL_SAVE_NAME)
    checkpoint = torch.load(checkpoint_path, weights_only=False)
    checkpoint['test_results'] = test_results
    checkpoint['optimal_threshold'] = test_results['optimal_threshold']
    checkpoint['training_history'] = history
    torch.save(checkpoint, checkpoint_path)

    print(f"\n✅ Two-stage training & evaluation complete!")
    print(f"   Model saved at: {checkpoint_path}")

    # ─────────────────────────────────────────────────────────
    # Summary
    # ─────────────────────────────────────────────────────────
    print("\n" + "=" * 70)
    print("TWO-STAGE TRAINING SUMMARY")
    print("=" * 70)

    phase1_epochs = min(model_cfg.FREEZE_EPOCHS, len(history['train_loss']))
    phase2_epochs = max(0, len(history['train_loss']) - model_cfg.FREEZE_EPOCHS)

    print(f"""
Model: {model_cfg.MODEL_NAME}
Image Size: {config.IMAGE_SIZE}x{config.IMAGE_SIZE}
─────────────────────────────────────────────────

Data:
  Train images: {len(train_df)}
  Val images:   {len(val_df)}
  Test images:  {len(test_df)}

Two-Stage Training:
  Phase 1 (Frozen backbone):
    Epochs:         {phase1_epochs}
    Learning rate:  {model_cfg.LR}
    BatchNorm:      {'Frozen' if model_cfg.FREEZE_BN else 'Trainable'}

  Phase 2 (Full fine-tuning):
    Epochs:         {phase2_epochs}
    Learning rate:  {model_cfg.LR * model_cfg.UNFREEZE_LR_FACTOR}

  Total epochs:     {len(history['train_loss'])}
  Batch size:       {config.BATCH_SIZE}
  Pos weight:       {model_cfg.POS_WEIGHT}
  Mixed precision:  {model_cfg.USE_MIXED_PRECISION}

Test Results (Threshold = 0.50):
  AUC:          {test_results['auc']:.4f}
  Accuracy:     {test_results['accuracy_default']:.4f}
  Precision:    {test_results['precision_default']:.4f}
  Recall:       {test_results['recall_default']:.4f}
  F1:           {test_results['f1_default']:.4f}

Test Results (Threshold = {test_results['optimal_threshold']:.2f}):
  Accuracy:     {test_results['accuracy_optimal']:.4f}
  Precision:    {test_results['precision_optimal']:.4f}
  Recall:       {test_results['recall_optimal']:.4f}
  F1:           {test_results['f1_optimal']:.4f}

─────────────────────────────────────────────────
""")

    # ─────────────────────────────────────────────────────────
    # Helper code for loading
    # ─────────────────────────────────────────────────────────
    print("\n" + "=" * 70)
    print("HELPER CODE FOR LOADING MODEL")
    print("=" * 70)

    helper_code = f'''
# Load trained EfficientNet-B2 model for multimodal fusion:

import torch
import torch.nn as nn
from torchvision import models

class BinaryDRClassifier(nn.Module):
    def __init__(self, model_name='efficientnet_b2', pretrained=False, dropout=0.3):
        super().__init__()
        self.backbone = models.efficientnet_b2(weights=None)
        n = self.backbone.classifier[1].in_features
        self.backbone.classifier = nn.Identity()
        self.feature_dim = n  # 1408 for B2
        self.classifier = nn.Sequential(
            nn.Dropout(dropout),
            nn.Linear(n, 256),
            nn.ReLU(),
            nn.BatchNorm1d(256),
            nn.Dropout(dropout * 0.5),
            nn.Linear(256, 1)
        )

    def extract_features(self, x):
        return self.backbone(x)

    def forward(self, x):
        f = self.backbone(x)
        return self.classifier(f)

def load_retinal_model(checkpoint_path, device='cpu'):
    checkpoint = torch.load(checkpoint_path, map_location=device)
    config = checkpoint.get('config', {{}})

    model = BinaryDRClassifier(
        model_name=config.get('model_name', 'efficientnet_b2'),
        pretrained=False,
        dropout=config.get('dropout', 0.3)
    )
    model.load_state_dict(checkpoint['model_state_dict'])
    model.to(device)
    model.eval()

    return model, checkpoint

# Usage:
# model, checkpoint = load_retinal_model('{model_cfg.MODEL_SAVE_NAME}')
# optimal_threshold = checkpoint['optimal_threshold']
# features = model.extract_features(images)  # Shape: (B, 1408)
'''

    print(helper_code)

    print("\n" + "=" * 70)
    print("✓ PIPELINE COMPLETE")
    print("=" * 70)
```
